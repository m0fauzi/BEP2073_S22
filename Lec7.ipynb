```python
import numpy as np
Ans = []
print('========================')
print('Time\tVelocity')
print('========================')
for t in range(5,30,2):
  if t <= 10:
    v = 11*t**2 - t
  elif t > 10 and t <= 20:
    v = 50*t + 2*(t-3)**2
  else:
    v = 3*np.exp(-3*t)
  Ans.append(v)
  print('%d\t%0.3f'%(t,v))
print('========================')
print(Ans)
```

```python
# Solve
# 10  + 21  = 1
# 30  + 51  = 2

A = np.array([[1,2],[3,5]])
B = np.array([1,2])
x = np.linalg.solve(A,B)
print('Answer:')
print('x0 =',round(x[0],3))
print('x1 =',round(x[1],3))
```
