```python
# Q1 Lec 6.1

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('dataQ1c.csv')

T = data.Temp
P = data.Press

n = 1000/(14+14)
V = 10

# PV = R * nT
# Y  = m * X
Y = P*V
X = T*n

for i,x in enumerate(X):
  if i == 1 or i == 4:
    plt.plot(x,Y[i],'ks')
    s = '('+str(round(x,1))+','+str(round(Y[i],1))+')'
    plt.text(x+1,Y[i]-0.5,s)
  else:  
    plt.plot(x,Y[i],'rs')

plt.show()

R = (Y[4] - Y[1]) /(X[4] - X[1])
print('Gas constant, R = %0.3f'%R)

```

```python
def V(h,R=3):
  pi = 22/7
  return pi*h**2*(3*R-h)/3


h = 0.01
Vcal = V(h)
for i in range(10000):
  e = 30 - Vcal
  if e > -0.01 and e < 0.01:
    break;
  else:
    h = h + 0.001
    Vcal = V(h)

print('#',i,'-> Answer, h = %0.3fm'%h)
```

```python
import scipy.optimize as so

def f(h,R=3):
  pi = 22/7
  V = 30
  return pi*h**2*(3*R-h)/3 - V

res = so.root_scalar(f,bracket = [-100,100], method='brentq')
print('Answer, h = %0.3fm'%res.root)
```
