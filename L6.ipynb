```python
X = [51,82,22,54,71,78,48,34,48,10]
sum = 0
# How to know number is odd or even
51%2 == 1  # 0 is for even number and 1 is for odd number

for x in X:
  if(x%2 == 1):
    print('Odd Number')
    sum = sum + x
  else:
    print('Even Number')
print(sum)


for i,x in enumerate(X):

  if(x%2 == 1):
    print('iter = %d; x = %d is Odd Number'%(i,x))
    sum = sum + x
  else:
    print('iter = %d; x = %d is Even Number'%(i,x))

print('Total odd number in X is ' + str(sum))
```
